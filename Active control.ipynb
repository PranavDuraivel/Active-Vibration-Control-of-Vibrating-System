```python
import numpy as np
import matplotlib.pyplot as plt

# System parameters
m = 5.0            # kg
k = 2.2e6          # N/m
c = 20.0           # Ns/m

# Frequency range
f = np.logspace(1, 3, 1000)  # frequency in Hz
w = 2 * np.pi * f            # convert to rad/s

# Transfer function G(j)
G = 1 / (-m * w**2 + 1j * c * w + k)

# Magnitude and Phase
mag = 20 * np.log10(np.abs(G))
phase = np.angle(G, deg=True)


# Plotting Bode plot with resonance marker
plt.figure(figsize=(12, 10))

# Magnitude plot
plt.subplot(2, 1, 1)
plt.semilogx(f, mag, label='Magnitude')
plt.title('Bode Plot of Open-Loop System')
plt.ylabel('Magnitude (dB)')
plt.grid(True, which="both", ls="--")
plt.legend()

# Phase plot
plt.subplot(2, 1, 2)
plt.semilogx(f, phase, label='Phase')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Phase (degrees)')
plt.grid(True, which="both", ls="--")
plt.legend()

plt.tight_layout()
plt.show()
```
